```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('/kaggle/input/train-cleaned/train.csv')
train_df.head()
```

```python
train_df['Total_Sales'] = train_df.luxury_sales + train_df.fresh_sales + train_df.dry_sales
train_df['Luxury_Sales_Ratio'] = train_df.luxury_sales/train_df.Total_Sales
train_df['Fresh_Sales_Ratio'] = train_df.fresh_sales/train_df.Total_Sales
train_df['Dry_Sales_Ratio'] = train_df.dry_sales/train_df.Total_Sales
train_df.head()
```

```python
outlet_cities_A = ['Batticaloa', 'Colombo', 'Jaffna', 'Gampaha', 'Kalmunai', 'Galle', 'Katunayake', 'Negombo', 'Trincomalee', 'Kandy', 'Anuradhapura']
outlet_cities_B = ['Kelaniya', 'Moratuwa', 'Wattala', 'Homagama', 'Dehiwala-Mount Lavinia', 'Panadura', 'Kaduwela', 'Peliyagoda', 'Kotte', 'Nuwara Eliya', 'Madawachiya']

def assign_outlet_city(x):
    if x in outlet_cities_A:
        return 1
    elif x in outlet_cities_B:
        return 0
    else:
        print("none")
        return None 
```

```python
train_df['outlet_city'] = train_df['outlet_city'].apply(assign_outlet_city)
train_df.head()
```

```python
train_df.drop(['Unnamed: 0', 'Customer_ID','luxury_sales','fresh_sales','dry_sales','Total_Sales'],axis=1,inplace=True)
train_df.head()
```
